```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('C:\\Users\\Mangavalli\\Documents\\GitHub\\COVID-19-and-its-Impact-on-Students-Data-Analysis\\COVID-19 Survey Student Responses.csv')

data.head()
```

```python
data.describe()
```

```python
average_age = data['Age of Subject'].mean()
print(f"1. Average Age of Participants: {average_age:.2f}")
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data['Age of Subject'],kde=True)
plt.title("Distribution of Participant Ages")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

According to the graph above, we see that the average age is roundly near the age of 20.

```python
rating_mapping = {'Excellent': 5, 'Good': 3, 'Very good': 4, 'Poor': 2, 'Very poor': 1}

data['Rating Numerical'] = data['Rating of Online Class experience'].map(rating_mapping)

correlation_online_class = data['Time spent on Online Class'].corr(data['Rating Numerical'])
print(f"Correlation between Time Spent on Online Classes and Rating: {correlation_online_class:.2f}")
```

```python
sns.regplot(x='Time spent on Online Class', y='Rating Numerical', data=data)
plt.title("Correlation between Time Spent on Online Classes and Rating")
plt.xlabel("Time Spent on Online Classes")
plt.ylabel("Rating Numerical")
plt.show()
```

Based on the graph above we can see that there is no correlation at all, and the calculated correlation coefficient is 0.15, which means that there is no correlation at all.

```python
weight_mapping = {'Increased': 1, 'Decreased': -1, 'Remain Constant': 0}
data['Weight Change Numerical'] = data['Change in your weight'].map(weight_mapping)


correlation_coefficient_weight_fitness = data['Time spent on fitness'].corr(data['Weight Change Numerical'])
print(f"10. Correlation between Change in Weight and Time Spent on Fitness: {correlation_coefficient_weight_fitness:.2f}")
```

```python
plt.figure(figsize=(10, 6))
sns.regplot(x='Time spent on fitness', y='Weight Change Numerical', data=data)
plt.title("Correlation between Time Spent on Fitness and Change in Weight")
plt.xlabel("Time Spent on Fitness")
plt.ylabel("Weight Change Numerical")
plt.show()
```

According to the plot above, we can see that there is no correlation between the time spent on fitness and weight change. The plot can be backed up by the correlation coefficient of -0.12, meaning there is no correlation.
